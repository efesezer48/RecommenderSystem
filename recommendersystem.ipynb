```python
import matplotlib.pyplot as plt
import seaborn as sns
categorical_columns = ['carrier', 'devicebrand']

# Selecting the top 10 most occurring categories from 'carrier' and 'devicebrand' columns
top_n = 10

fig, axs = plt.subplots(1, 2, figsize=(18, 9))

for i, column in enumerate(categorical_columns):
    value_counts = data[column].value_counts().nlargest(top_n)
    axs[i].pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'Top {top_n} {column} kategorisi')

plt.tight_layout()
plt.show()
```

```python
def binary_function(rows):
    binary_menu = np.zeros(9).astype(int)
    for row in rows.split(", "):
        menu_id = int(row.replace("menu", "")) - 1
        
        binary_menu[menu_id] = 1
        
    return binary_menu
data["Binary_target"] = data["target"].apply(binary_function)

string_binary_target = ["".join(map(str, arr)) for arr in data["Binary_target"].values]
data["Binary_target"] = string_binary_target
data
```

```python
data = data.drop(["id","month", "target","carrier","devicebrand"], axis=1)
```

```python
#x = data.iloc[:,53].values.reshape(-1,1)
x = data.drop(['n_seconds_1','n_seconds_2','n_seconds_3'],axis=1)
y = data.iloc[:,53].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix as cm    
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=80)
rf = RandomForestClassifier(n_estimators=100, max_depth = 3, random_state=42)
rf.fit(X_train, y_train)
```

```python
predictions = rf.predict(X_test)
score = round(accuracy_score(y_test, predictions), 3)
cm1 = cm(y_test, predictions)
sns.heatmap(cm1, annot=True, fmt=".0f")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Accuracy Score: {0}'.format(score), size = 15)
plt.show()
```
